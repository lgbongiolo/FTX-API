```python
import pandas as pd
import requests
import datetime
import plotly.graph_objects as go

```

```python
start = "2017-01-01"
end = "2020-01-01" #datetime.datetime.now()


historical = requests.get('https://ftx.com/api/markets/BTC-0930/candles?resolution=3600&start_time=1609462800').json()
historical = pd.DataFrame(historical['result'])
historical.drop(['startTime'], axis = 1, inplace=True)
historical['time'] = pd.to_datetime(historical['time'], unit='ms')
historical.set_index('time', inplace=True)
historical.head(30)
```

```python

fig = go.Figure(data=[go.Candlestick(x = historical.index,
                                    open = historical['open'],
                                    high = historical['high'],
                                    low = historical['low'],
                                    close = historical['close'],
                                    ),
                     go.Scatter(x=historical.index, line=dict(color='purple', width=1))])

fig.show()

```
